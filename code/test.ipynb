```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import random
```

```python
data_file_path = '/Users/davidan/Desktop/Github/cogs9_spotify_study/data/'
file_name = 'Spotify_2000.csv'
spotify_2000_df = pd.read_csv(data_file_path + file_name)
spotify_2000_df
```

```python
s = spotify_2000_df
replace = lambda x: x.replace(',', '')
length_int = s.get('Length (Duration)').apply(replace).apply(int)
trimmed_spotify_2000_df = pd.DataFrame().assign(**{'Title': s.get('Title'), 'Artist': s.get('Artist'), \
                                                   'Year': s.get('Year'), 'Length': length_int, \
                                                   'Popularity': s.get('Popularity')})
```

```python
trimmed_spotify_2000_df.plot.scatter(x='Length', y='Popularity', s=0.2, title='Relationship between song length and popularity').set_xlabel('Length (sec)')
```

```python
year_range = np.arange(2001, 2020)
for y in year_range:
    by_year = trimmed_spotify_2000_df[trimmed_spotify_2000_df.get('Year') == y]
    by_year.plot.scatter(x='Length', y='Popularity', s=10, title='Relationship between song length and popularity (%s)'%(str(y)), yticks=np.arange(0,101,20), xticks=np.arange(0,901,100)).set_xlabel('Length (sec)')
```
